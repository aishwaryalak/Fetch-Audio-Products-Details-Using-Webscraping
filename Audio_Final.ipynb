```python
Product_name=[]
Flipkart_Price = []
Original_price=[]
Disc=[]
Model=[]
Rating=[]
Colour=[]
Headphone=[]
Remote=[]
Sales_pack=[]
Connect=[]
Reviews=[]
for i in l:
    page=requests.get(i)
    soup=BeautifulSoup(page.content)
    
    #TITLE OF THE PRODUCT
    title=soup.title
    print(title)
    
    #Product Name
    product = soup.find_all("span",class_="B_NuCI")
    print(product)
    #converting product name to list
    for a in range(0,len(product)):
        Product_name.append(product[a].get_text())
        print(Product_name)
        
    #Original Price
    ori = soup.find_all("div",class_="_3I9_wc _2p6lqe")
    print(product)
    #converting original price to list
    for d in range(0,len(ori)):
        Original_price.append(ori[d].get_text())
        print(Original_price)
        
    #Flipkart price
    Price = soup.find_all("div",class_="_30jeq3 _16Jk6d")
    print(Price)
    #price in list format
    for b in range(0,len(Price)):
        Flipkart_Price.append(Price[b].get_text())
        print(Flipkart_Price)
        
    #Discount
    disc = soup.find_all("div",class_="_3Ay6Sb _31Dcoz")
    print(disc)
    for j in range(0,len(disc)):
        Disc.append(disc[j].get_text())
        print(Disc)
        
    #Rating
    rat = soup.find_all("div",class_="_2d4LTz")
    print(rat)
    for c in range(0,len(rat)):
        Rating.append(rat[c].get_text())
        print(Rating)
        
        
    #Reviews
    rev = soup.find("p",class_="_2-N8zT")
    print(rev)
    Reviews.append(rev.text)
        
        
    #Specifications
    spec=soup.find('table', attrs={'class':'_14cfVK'})
    print(spec)
    for each in spec:
            col=each.find_all('li', attrs={'class':'_21lJbe'})
            model=col[0].text
            colour = col[1].text
            headphone = col[2].text
            inline_r = col[3].text
            s_package = col[4].text
            conn = col[5].text
            #append in form of a lsit
            Model.append(model)
            Colour.append(colour)
            Headphone.append(headphone)
            Remote.append(inline_r)
            Sales_pack.append(s_package)
            Connect.append(conn)
            
```
